####Potential cleaning tasks:
- Convert date columns (Order_Date, Ship_Date) to datetime format.
Check for duplicates.
Normalize categorical values (e.g., standardizing text formatting).
Identify and remove any outliers in the Sales column.

```python
# Convert date columns to datetime format
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], errors='coerce')

# Check for duplicate rows
duplicate_rows = df.duplicated().sum()

# Check for outliers in Sales using IQR
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['Sales'] < lower_bound) | (df['Sales'] > upper_bound)]

# Standardize categorical columns (strip whitespace and capitalize properly)
categorical_cols = ['Ship_Mode', 'Segment', 'City', 'State', 'Region', 'Category', 'Sub_Category', 'Product_Name']
df[categorical_cols] = df[categorical_cols].apply(lambda x: x.str.strip().str.title())

# Summary of cleaning actions
duplicate_rows, len(outliers)
```

Data Cleaning Summary:
- Date Columns: Converted Order_Date and Ship_Date to datetime format.
- Duplicates: Found and will remove 1 duplicate row.
- Outliers in Sales: Detected 1,145 outliers using the IQR method.
- Categorical Standardization: Stripped whitespace and applied proper capitalization.

```python
# Cap the outliers within the IQR range
df['Sales'] = df['Sales'].clip(lower=lower_bound, upper=upper_bound)

# Remove duplicate row
df = df.drop_duplicates()

# Verify that outliers are handled and duplicates removed
df['Sales'].describe(), df.duplicated().sum()
```

```python
df['Sales'] = df['Sales'].astype(int)

# Change the type of 'Order_Date' and 'Ship_Date'
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])

df['year_order_date'] = df['Order_Date'].dt.year
df['month_order_date'] = df['Order_Date'].dt.month
df['weekday_order_date'] = df['Order_Date'].dt.weekday

df['year_ship_date'] = df['Ship_Date'].dt.year
df['month_ship_date'] = df['Ship_Date'].dt.month
df['weekday_ship_date'] = df['Ship_Date'].dt.weekday
```

```python
# Separate features (X) and target variable (y)
X = df.drop(columns=["Sales"])  # Features
y = df["Sales"]  # Target variable
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Convert date columns to numeric (if necessary)
date_columns = ["year_order_date", "month_order_date", "weekday_order_date",
                "year_ship_date", "month_ship_date", "weekday_ship_date"]

# Ensure these columns exist in the dataset
for col in date_columns:
    if col in X.columns:
        X[col] = X[col].astype(int)  # Convert datetime to integer (year/month already int)

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Select only numerical features for scaling
numerical_features = X.select_dtypes(include=np.number).columns
X_numeric = X[numerical_features]

# Fit and transform only the numerical features (X_numeric)
X_scaled = scaler.fit_transform(X_numeric)

# Convert back to DataFrame
X_scaled = pd.DataFrame(X_scaled, columns=X_numeric.columns, index=X.index) # Keep original index

# If you need the date columns back in the scaled DataFrame:
X_scaled = pd.concat([X_scaled, X[date_columns]], axis=1)

# Display scaled features
print(X_scaled.head())
```

- MinMaxScaler scales all values to a range of [0,1].
- X_scaled now contains normalized features.
- We do not scale y (Sales) unless needed for deep learning.

```python
# Scale the target variable (Sales)
y_scaler = MinMaxScaler()
y_scaled = y_scaler.fit_transform(y.values.reshape(-1, 1))

# Convert back to a DataFrame (optional)
y_scaled = pd.DataFrame(y_scaled, columns=["Sales"])

# To inverse transform later (after prediction)
y_original = y_scaler.inverse_transform(y_scaled)

```
