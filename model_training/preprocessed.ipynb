```python
# Drop the Postal_Code column
df.drop(columns=["Postal_Code"], inplace=True)
```

```python
# Drop the 'Country' column
df_cleaned = df.drop(columns=["Country"],inplace=True)
```

```python
# Removing unnecessary columns
columns_to_drop = ["Row_ID", "Order_ID", "Customer_ID", "Product_ID", "Customer_Name"]

# Removing unnecessary columns in the original dataset
df.drop(columns=columns_to_drop, inplace=True)

# Summarizing the dataset
df.info()
```

```python
# Displaying the first 5 rows of the cleaned dataset
df.head()
```

```python
# Converting date columns to datetime format
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], errors='coerce')
```

```python
# Checking for duplicate rows
duplicate_rows = df.duplicated().sum()
print(duplicate_rows)
```

```python
# Removing duplicate row
df = df.drop_duplicates()
dr=df.duplicated().sum()
print(dr)
```

```python
# Checking for outliers in Sales
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

# Lower bound:
lower_bound = Q1 - 1.5 * IQR

# Upper bound:
upper_bound = Q3 + 1.5 * IQR

# Finding outliers
outliers = df[(df['Sales'] < lower_bound) | (df['Sales'] > upper_bound)]

# Standardizing categorical columns
categorical_cols = ['Ship_Mode', 'Segment', 'City', 'State', 'Region', 'Category', 'Sub_Category', 'Product_Name']
df[categorical_cols] = df[categorical_cols].apply(lambda x: x.str.strip().str.title())

# Summary of cleaning actions
len(outliers)

```

```python
# Any values below the lower bound are set to the lower bound
# Any values above the upper bound are set to the upper bound
df['Sales'] = df['Sales'].clip(lower=lower_bound, upper=upper_bound)


# Displaying summary of the 'Sales' column
df['Sales'].describe()
```

```python
df.head() # Printing 1st 5 rows of dataframe
```

```python
# Convert the 'Sales' column values to integers for consistency in numerical operations
df['Sales'] = df['Sales'].astype(int)
```

```python
# Convert 'Order_Date' and 'Ship_Date' columns from string/object format to DateTime format
# This allows easy extraction of date-related information and facilitates time-based analysis
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])
```

```python
# Extract the year, month, and weekday from 'Ship_Date' to create new columns
df['year_ship_date'] = df['Ship_Date'].dt.year  # Extracts the year (e.g., 2023)
df['month_ship_date'] = df['Ship_Date'].dt.month  # Extracts the month (1-12)
df['weekday_ship_date'] = df['Ship_Date'].dt.weekday  # Extracts the weekday (0 = Monday, 6 = Sunday)
```

```python
# Extract the year, month, and weekday from 'Order_Date' to create new columns
df['year_order_date'] = df['Order_Date'].dt.year  # Extracts the year (e.g., 2023)
df['month_order_date'] = df['Order_Date'].dt.month  # Extracts the month (1-12)
df['weekday_order_date'] = df['Order_Date'].dt.weekday  # Extracts the weekday (0 = Monday, 6 = Sunday)
```
